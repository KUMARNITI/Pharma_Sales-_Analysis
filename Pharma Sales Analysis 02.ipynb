```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for two different datasets
categories = ['Jan23', 'Feb23', 'Mar23', 'Apr23','May23','Jun23']
data1 = [120019, 280142, 360566, 400081,520272,639257]  # Data for the first dataset
data2 = [239963,240135,396948,440184,600150,719316]   # Data for the second dataset

x = np.arange(len(categories))  # X-axis positions for the bars
bar_width = 0.4  # Width of each bar

# Create a new figure and set the size
plt.figure(figsize=(10, 6))

# Plot the bars for the first dataset
plt.bar(x - bar_width/2, data1, width=bar_width, align='center', label='Fludara_dataset')

# Plot the bars for the second dataset
plt.bar(x + bar_width/2, data2, width=bar_width, align='center', label='Mercapto_dataset')

# Set labels, title, and legend
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Clustered Bar Graph of Two Different Datasets')
plt.xticks(x, categories)
plt.legend()

# Show the clustered bar graph
plt.tight_layout()
plt.show()
```
